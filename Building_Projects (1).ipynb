# 1. [Human Anatomy] Quiz Game
*__Could definitely be any subject__*

```python
import re
from word2number import w2n
def normalize_answer(answer): #this function ensures that user input is recognized irrespective of space, article or word/number
    answer = answer.lower().strip()
    answer = re.sub(r'\b(a|an|the)\b', '', answer)
    answer = re.sub(r'\s+', '', answer).strip()
    try:
        answer = str(w2n.word_to_num(answer))
    except:
        pass
    return answer
    
print("Welcome to the Anatomy Wizard Quiz")
playing = input("Are you ready to play? ").lower()
if playing != "yes":
    quit()
#initialize test score
score = 0
print("Great! Let's Dive In")

#Question 1
answer = input("What is the largest organ in the human body? ")
if normalize_answer(answer) == normalize_answer("skin"):
    print("That is Correct!")
    score += 1
else:
    print("Nah, You're Wrong!")

#Question 2
answer = input("How many chambers are there in the human heart? ")
if normalize_answer(answer) == normalize_answer("4"):
    print("That is Correct!")
    score += 1
else:
    print("Nah, You're Wrong!")

#Question 3
answer = input("Which bone is commonly known as the collarbone? ")
if normalize_answer(answer) == normalize_answer("clavicle"):
    print("That is Correct!")
    score += 1
else:
    print("Nah, You're Wrong!")

print(f"Your Final score is: {score}")
print("That is: " + str((score/3) * 100) + "%")
```

# 2. Number Guessing Game

```python
import random
import time


def get_top_of_range():
    difficulty = input("Choose difficulty (easy, medium, hard): ").lower()
    if difficulty == "easy":
        return 10
    elif difficulty == "medium":
        return 50
    elif difficulty == "hard":
        return 100
    else:
        print("Invalid choice, defaulting to easy.")
        return 10


def get_user_guess():
    while True:
        guess = input("Make a guess: ")
        if guess.isdigit():
            return int(guess)
        else:
            print("Please enter a valid number.")


def give_feedback(user_guess, random_number):
    difference = abs(user_guess - random_number)
    if user_guess > random_number:
        print("Too high!")
    else:
        print("Too low!")

    if difference <= 3:
        print("You're very close!")
    elif difference <= 10:
        print("You're getting warmer.")


def play_game():
    top_of_range = get_top_of_range()
    random_number = random.randint(1, top_of_range)
    guesses = 0
    max_guesses = 10

    print(f"\nI'm thinking of a number between 1 and {top_of_range}. You have {max_guesses} guesses.")

    while guesses < max_guesses:
        guesses += 1
        user_guess = get_user_guess()

        if user_guess == random_number:
            print(f"🎉 You got it in {guesses} guesses! 🎉\n")
            return guesses

        give_feedback(user_guess, random_number)
        print(f"You have {max_guesses - guesses} guesses left.\n")

    print(f"\n😞 You've run out of guesses! The number was {random_number}.\n")
    return None


def main():
    best_score = None
    print("Welcome to the Number Guessing Game!")
    time.sleep(1)

    while True:
        result = play_game()
        if result:
            if best_score is None or result < best_score:
                best_score = result
                print("🏆 New best score!")

        play_again = input("Do you want to play again? (yes/no): ").lower()
        if play_again != "yes":
            break

    print("\nThanks for playing!")
    if best_score:
        print(f"Your best score was {best_score} guesses.")
    print("Goodbye 👋")


if __name__ == "__main__":
    main()
```
